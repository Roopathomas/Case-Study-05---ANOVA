# ● Check whether all the zones generate the same amount of sales.

# 
**Null hypothesis: all the zones generate the same amount of sales.**
<br>
**Alternate hypothesis: Not all the zones generate the same amount of sales.**
<br>
**Now we can perform a oneway analysis of variance using α=0.05**
</p>

```python
#import scipy 
from scipy import stats
fvalue, pvalue = stats.f_oneway(Data['Zone - A'], Data['Zone - B'], Data['Zone - C'],Data['Zone - D'])
print('f-value:',round(fvalue,4),'p-value:',round(pvalue,4))
if pvalue <0.05:    # alpha value is 0.05 or 5%
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```

# Insight:
**Here we reject the null hypothesis because p value is less than 0.05.**
<br>
**Therefore,Not all the zones generate the same amount of sales.**
